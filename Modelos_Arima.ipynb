# **series de tiempo**

# **Conceptos fundamentales**

 **1. Procesos estoc√°sticos**
 
Un proceso estoc√°stico o aleatorio es una colecci√≥n de variables aleatorias ordenadas en el
tiempo. Si $Y$ denota una variable aleatoria y es continua, se denota como $Y(t)$, pero si es discreta
se expresa como $Y_t$.
se dice que un proceso estoc√°stico es estacionario si su media y su varianza son constantes en el tiempo y si el valor de la covarianza entre dos periodos depende s√≥lo de la distancia
o rezago entre estos dos periodos, y no del tiempo en el cual se calcul√≥ la covarianza.

**2. Proceso estoc√°stico de ra√≠z unitaria**

es un tipo de proceso estoc√°stico (aleatorio) que exhibe una caracter√≠stica espec√≠fica: un solo impacto o choque puede tener un efecto persistente en el proceso, influyendo en √©l indefinidamente

**3. Pruebas de estacionariedad**

dos pruebas: 1) el an√°lisis gr√°fico y 2) la prueba del correlograma. 

**4. Prueba de ra√≠z unitaria**

Una prueba de ra√≠z unitaria es un procedimiento estad√≠stico utilizado para determinar si una serie temporal es estacionaria o no estacionaria.

prueba Dickey-Fuller (DF):

Hip√≥tesis nula: $H_0:Œ¥$ = 0 (es decir, existe una ra√≠z unitaria, la serie de tiempo es no estacionaria o tiene tendencia estoc√°stica). 

Hip√≥tesis alternativa: $H_1:Œ¥$ < 0 (es decir, la serie de tiempo es estacionaria, posiblemente
alrededor de una tendencia determinista)

**5. Transformaci√≥n de las series de tiempo no estacionarias**

Para evitar el problema de la regresi√≥n espuria que pudiese surgir al hacer la regresi√≥n de una serie de tiempo no estacionaria sobre una o m√°s series de tiempo no estacionarias tenemos que transformar las series de tiempo no estacionarias en
estacionarias. 

El m√©todo de transformaci√≥n depende de que las series de tiempo sean procesos
estacionarios en diferencias (PED) o procesos estacionarios con tendencia (PET).

Procesos estacionarios en diferencias:

Si una serie de tiempo tiene una ra√≠z unitaria, las primeras diferencias de tales series son estacionarias. En consecuencia, la soluci√≥n aqu√≠ es tomar las primeras diferencias de las series de tiempo.

Procesos estacionarios en tendencia:

Un PET es estacionario alrededor de la l√≠nea de tendencia. Por tanto, la manera m√°s sencilla de convertir en estacionaria una serie de tiempo es hacer la regresi√≥n de ella sobre el tiempo y los residuos de tal regresi√≥n ser√°n estacionarios. 

# **Creaci√≥n de modelos AR, MA y ARIMA para series de tiempo**


La publicaci√≥n de G. P. E. Box y G. M. Jenkins Time Series Analysis: Forecasting and Control, op. cit., marc√≥ el comienzo de una nueva generaci√≥n de herramientas de pron√≥stico. Popularmente conocida como metodolog√≠a de Box-Jenkins (BJ), pero t√©cnicamente conocida como metodolog√≠a ARIMA, el inter√©s de estos m√©todos de pron√≥sticos no est√° en la construcci√≥n de
modelos uniecuacionales o de ecuaciones simult√°neas, sino en el an√°lisis de las propiedades probabil√≠sticas, o estoc√°sticas, de las series de tiempo econ√≥micas por s√≠ mismas seg√∫n la filosof√≠a de que los datos hablen por s√≠ mismos. A diferencia de los modelos de regresi√≥n, en los cuales Yt
se explica por las k regresoras X1, X2, X3, . . . , Xk, en los modelos de series de tiempo del tipo BJ,
Yt se explica por valores pasados o rezagados de s√≠ misma y por los t√©rminos de error estoc√°sticos. Por esta raz√≥n, los modelos ARIMA reciben algunas veces el nombre de modelos ate√≥ricos
‚Äîporque no se derivan de teor√≠a econ√≥mica alguna‚Äî, y las teor√≠as econ√≥micas a menudo son la
base de los modelos de ecuaciones simult√°neas.


Una serie de tiempo es una secuencia en la que se registra una m√©trica en intervalos de tiempo regulares.
* anual (por ejemplo, presupuesto anual), 
* trimestral (por ejemplo, gastos), 
* mensual (por ejemplo, tr√°fico a√©reo), 
* semanal (por ejemplo, cantidad de ventas), 
* diaria (por ejemplo, clima), 
* por hora. (p. ej., precio de las acciones), 
* minutos (p. ej., llamadas entrantes en un galope de llamadas) e incluso 
* segundos (p. ej., tr√°fico web).

Un modelo ARIMA se caracteriza por 3 t√©rminos: p, d, q d√≥nde,

* p: es el orden del t√©rmino AR

* q: es el orden del t√©rmino MA

* :d es el n√∫mero de diferenciaciones necesarias para que la serie de tiempo sea estacionaria

'p' es el orden del t√©rmino 'Auto regresivo' (AR). Se refiere al n√∫mero de rezagos de $Y$ que se utilizar√°n como predictores. Y 'q' es el orden del t√©rmino 'Media m√≥vil' (MA). Se refiere al n√∫mero de errores de pron√≥stico retrasados que deben incluirse en el modelo ARIMA.
El valor de d, por lo tanto, es el n√∫mero m√≠nimo de diferenciaciones necesarias para que la serie sea estacionaria. Y si la serie de tiempo ya es estacionaria, entonces $d$ = 0.

Si una serie de tiempo tiene patrones estacionales, entonces necesita agregar t√©rminos estacionales y se convierte en SARIMA, abreviatura de "Seasonal ARIMA". M√°s sobre eso una vez que terminemos ARIMA.


**4.1. Proceso autorregresivo AR ($p$)**

Un modelo auto regresivo puro (solo AR) es aquel en el que $Y_t$ depende solo de sus propios retrasos. Es decir, $Y_t$ es una funci√≥n de los 'rezagos de $Y_t$.

$$Y_t=\alpha+\beta_1 Y_{t-1}+\beta_2 Y_{t-2}+\cdots+\beta_p Y_{t-p}+u_t $$

$Y_t$: valor de la serie en el tiempo $t$

$ùõº$: constante o t√©rmino independiente

$ùõΩ_1,ùõΩ_2,‚Ä¶,ùõΩ_ùëù$ : coeficientes del modelo que indican la influencia de los valores pasados

$ùëå_{ùë°‚àí1},ùëå_{ùë°‚àí2},‚Ä¶,ùëå_{ùë°‚àíùëù}$ : valores rezagados de la serie (lags)

$u_t$: t√©rmino de error aleatorio (ruido blanco)

Operador rezago:

$$Y_ùë° ‚àí ùõΩ_1Y_{ùë°‚àí1} ‚àí ùõΩ_2Y_{ùë°‚àí2} ‚àí‚Ä¶‚àí ùõΩ_ùëùY_{ùë°‚àíùëù} = ùõº + ùë¢_ùë°$$

$$(1 ‚àí ùõΩ_1ùêø^1 ‚àí ùõΩ_2ùêø^2 ‚àí ‚ãØ ‚àí ùõΩ_ùëùùêø^ùëù)Y_ùë° = ùõº + ùë¢_ùë°$$

Modelos s√≥lo consideran los valores actuales

**4.2. Proceso de medias m√≥viles MA ($q$)**

Del mismo modo, un modelo puro de media m√≥vil (solo MA) es uno en el que $Y_t$ depende solo de los errores de pron√≥stico retrasados.

$$Y_t=\alpha+u_t+\phi_1 u_{t-1}+\phi_2 u_{t-2}+\cdots+\phi_q u_{t-q}$$

$$Y_ùë° = ùõº + ùë¢_ùë°(1 + \phi_1ùêø^1 + ‚ãØ + \phi_ùëûùêø^ùëû)$$

$Y_ùë°$ es la variable observada.

$ùõø$ es una constante (media).

$ùë¢_ùë°$ es un t√©rmino de error aleatorio (ruido blanco).

$\phi_ùëñ$ son los coeficientes del modelo MA.

$ùëû$ es el orden del modelo.

En resumen, un proceso de promedios m√≥viles es tan s√≥lo una combinaci√≥n lineal de t√©rminos de error de ruido blanco.

**4.3. Proceso autorregresivo y de promedios m√≥viles ARMA ($p$,$q$)**

es muy probable que $Y$ tenga caracter√≠sticas de AR y de MA a la vez, Y, por consiguiente, sea ARMA:

$$Y_ùë° = ùõº + ùõΩ_1Y_{ùë°‚àí1} + ùõΩ_2ùë¶_{ùë°‚àí2} +‚Ä¶+ùõΩ_ùëùùë¶_{ùë°‚àíùëù} + ùë¢_ùë° +\phi_1ùë¢_{ùë°‚àí1} + \phi_2ùë¢_{ùë°‚àí2} + ‚Ä¶ + \phi_ùëûùë¢_{ùë°‚àíùëû}$$

$Y_ùë°$: valor de la serie en el tiempo $t$

$ùõº$: constante o media

$ùõΩ_ùëñ$: coeficientes del componente AR (autoregresivo)

$ùúô_ùëó$: coeficientes del componente MA (media m√≥vil)

$ùë¢_ùë°$: ruido blanco (residual aleatorio)

**4.4 Proceso autorregresivo integrado de promedios m√≥viles ARIMA ($p$,$d$,$q$)**

Una serie de tiempo autorregresiva integrada de promedios m√≥viles, donde $p$ denota el n√∫mero de t√©rminos autorregresivos, $d$ el n√∫mero de veces que la serie debe diferenciarse para hacerse estacionaria y $q$ el n√∫mero de t√©rminos de promedios m√≥viles.

$$‚àÜ^ùíÖY_ùíï = ùú∂ + ùú∑_ùüè‚àÜ^ùíÖY_{ùíï‚àíùüè} + ùú∑_ùüê‚àÜ^ùíÖY_{ùíï‚àíùüê} +‚Ä¶+ùú∑_ùíë‚àÜ^ùíÖY_{ùíï‚àíùíë} + ùíñ_ùíï +\phi_ùüèùíñ_{ùíï‚àíùüè} + \phi_ùüêùíñ_{ùíï‚àíùüê} + ‚Ä¶ + \phi_ùííùíñ_{ùíï‚àíùíí}$$

$Œî^dY_t$ es la serie diferenciada d veces (para hacerla estacionaria).

$ùõº$ es la constante.

$ùõΩ_ùëñ$: coeficientes AR (autorregresivos).

$ùúô_ùëñ$: coeficientes MA (media m√≥vil).

$ùë¢_ùë°$: ruido blanco.

En forma simplificada

$$
‚àÜ^dYùíï = ùú∂ + \sum_{i=1}^{p} ùõΩ_ùíä‚àÜ^dY_{ùíï‚àíùíä} + ùíñ_ùíï +\sum_{i=1}^{p} \phi_ùíãùíñ_{ùíï‚àíùíã}

## **Metodolog√≠a BOX-JENKING**

El m√©todo considera cuatro pasos:

Paso 1. Identificaci√≥n. Es decir, encontrar los valores apropiados de p, d y q.
Las herramientas principales en la identificaci√≥n son la funci√≥n de autocorrelaci√≥n (FAC), la
funci√≥n de autocorrelaci√≥n parcial (FACP) y los correlogramas resultantes, que son simplemente los gr√°fi cos de FAC y de FACP respecto de la longitud del rezago.

Paso 2. Estimaci√≥n. Tras identificar los valores apropiados de p y q, la siguiente etapa es
estimar los par√°metros de los t√©rminos autorregresivos y de promedios m√≥viles incluidos en
el modelo. 

Paso 3. Examen de diagn√≥stico. Despu√©s de seleccionar un modelo ARIMA particular y
de estimar sus par√°metros, tratamos de ver si el modelo seleccionado se ajusta a los datos en
forma razonablemente buena, pues es posible que exista otro modelo ARIMA que tambi√©n
lo haga.

Paso 4. Pron√≥stico. Una raz√≥n de la popularidad del proceso de construcci√≥n de modelos
ARIMA es su √©xito en el pron√≥stico. En muchos casos, los pron√≥sticos obtenidos por este
m√©todo son m√°s confi ables que los obtenidos de modelos econom√©tricos tradicionales, en
particular en el caso de pron√≥sticos de corto plazo.

## **Errores de pronostico**

Desviaci√≥n absoluta de la media (DAM):

$$
DAM = \frac{\sum\limits_{t=1}^{n} \left| Y_t - \hat{Y}_t \right|}{n}
$$

Error medio cuadratico (EMC):

$$
EMC = \frac{\sum\limits_{t=1}^{n} (Y_t - \hat{Y}_t)^2}{n}
$$

% Error medio absoluto (PEMA):

$$
PEMA = \frac{\sum\limits_{y=1}^{n}  \frac{\left|Y_t - \hat{Y}_t\right|}{Y_t}}{n}
$$

% medio de error (PME):

$$
PME = \frac{\sum\limits_{t=1}^{n}  \frac{(Y_t - \hat{Y}_t)}{Y_t}}{n}
$$

## **5. C√≥mo encontrar el orden de diferenciaci√≥n (d) en el modelo ARIMA**
El prop√≥sito de diferenciarlo para hacer estacionaria la serie de tiempo.


Primero, vamos a verificar si la serie es estacionaria usando la **prueba Augmented Dickey Fuller**, del paquete `statsmodels.adfuller()`

¬øPor qu√©?

Porque solo se necesita diferenciar si la serie no es estacionaria. De lo contrario, no se necesita diferenciaci√≥n, es decir, $d = 0$.

La hip√≥tesis nula de la prueba ADF es que la serie de tiempo no es estacionaria. Entonces, si el valor $p$ de la prueba es menor que el nivel de significancia (0.05), entonces rechaza la hip√≥tesis nula e infiere que la serie de tiempo es de hecho estacionaria.

Entonces, en nuestro caso, si $P- Value> 0.05$ seguimos adelante con la b√∫squeda del orden de diferenciaci√≥n.

Graficos la serie para identificar si presenta estacionaridad y estacionalidad

Hacemos la prueba de `Dickey-Fuller` para probar si la serie presenta estacionaridad

```python
from statsmodels.tsa.stattools import adfuller

# Suponiendo que 'df' tiene la columna 'Close' y el √≠ndice es la fecha
serie = df['ventas'].dropna()

# Prueba ADF
resultado = adfuller(serie)

# Mostrar resultados
print('Estad√≠stico ADF:', resultado[0])
print('Valor-p:', resultado[1])
print('Lags utilizados:', resultado[2])
print('N√∫mero de observaciones:', resultado[3])
print('Valores cr√≠ticos:')
for clave, valor in resultado[4].items():
    print(f'   {clave}: {valor}')
```

```python
plt.rcParams.update({'figure.figsize':(12,15), 'figure.dpi':120})
# Serie Original
fig, axes = plt.subplots(2, 2, sharex=False)
axes[0, 0].plot(df,color="r"); axes[0, 0].set_title('Serie Original')
plot_acf(df, ax=axes[0, 1], color="r")

# Primera diferenciaci√≥n 
axes[1, 0].plot(df.diff(), color="g"); axes[1, 0].set_title("Diferenciaci√≥n de Primer orden")
plot_acf(df.diff().dropna(), ax=axes[1, 1],color="g")
plt.show();
```

## **C√≥mo encontrar el orden del t√©rmino AR (p)**
El modelo necesita alg√∫n t√©rmino de AR. Puede averiguar el n√∫mero requerido de t√©rminos AR inspeccionando el gr√°fico de Autocorrelaci√≥n parcial (PACF).

La autocorrelaci√≥n parcial se puede imaginar como la correlaci√≥n entre la serie y su rezago, despu√©s de excluir las contribuciones de los rezagos intermedios. Entonces, PACF transmite la correlaci√≥n pura entre un retraso y la serie. De esa manera, sabr√° si ese retraso es necesario en el plazo de AR o no.


Puede observar que el retraso PACF 1 es bastante significativo ya que est√° muy por encima de la l√≠nea de importancia. El retardo 2 y 3 tambi√©n resulta significativo, logrando cruzar levemente el l√≠mite de significancia (regi√≥n azul). Pero voy a ser conservador y tentativamente fijar√© la p como 1.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Supongamos que tu DataFrame es `ventas_diff1` de la serie diferenciada y tienes una serie temporal en 'ventas'
serie = df1['ventas_diff1'].dropna()

# Crear gr√°ficos ACF y PACF
plt.figure(figsize=(14, 6))

# Autocorrelaci√≥n
plt.subplot(1, 2, 1)
plot_acf(serie, lags=30, ax=plt.gca())
plt.title('Autocorrelaci√≥n (ACF)')

# Autocorrelaci√≥n parcial
plt.subplot(1, 2, 2)
plot_pacf(serie, lags=30, ax=plt.gca(), method='ywm')  # method puede ser 'ywm' o 'ols'
plt.title('Autocorrelaci√≥n Parcial (PACF)')

plt.tight_layout()
plt.show()
```

no se identifico nigun AR

## **C√≥mo encontrar el orden del t√©rmino MA (q)**
De la misma forma que observamos el gr√°fico PACF para el n√∫mero de t√©rminos AR, puede mirar el gr√°fico ACF para el n√∫mero de t√©rminos MA. Un t√©rmino MA es t√©cnicamente el error del pron√≥stico retrasado.

El ACF indica cu√°ntos t√©rminos MA se requieren para eliminar cualquier autocorrelaci√≥n en la serie estacionarizada.

Veamos el gr√°fico de autocorrelaci√≥n de la serie diferenciada.

Un par de retrasos est√°n muy por encima de la l√≠nea de importancia. Entonces, fijemos tentativamente q como 1. En caso de duda, opte por el modelo m√°s simple que explique suficientemente la Y.

El termino MA se encontro q=1, Q=1

# **Paso 2. Estimaci√≥n**

## **Construir el modelo SARIMA**
Ahora que ha determinado los valores de (p, d=1, q=1), (P,D=1,Q=1) tiene todo lo necesario para ajustarse al modelo SARIMA. Usemos la implementaci√≥n en package SARIMAX()statsmodels

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# SARIMA(p,d,q)(P,D,Q)[s]
model = SARIMAX(df['ventas'], 
                order=(0,1,1),           # par√°metros ARIMA
                seasonal_order=(0,1,1,12), # estacionales: (P,D,Q,s)
                enforce_stationarity=False,
                enforce_invertibility=False)

result = model.fit(disp=False)
```

```python
print(result.summary())

# Visualizar pron√≥stico ajustado vs original
result.plot_diagnostics(figsize=(12, 6))
plt.show()
```

El resumen del modelo revela mucha informaci√≥n. La tabla en el medio es la tabla de coeficientes donde los valores bajo 'coef' son los pesos de los t√©rminos respectivos.

Observe aqu√≠ que el coeficientes del modelo es cercano a cero y el P-Valor es menor que 0.05.


# **Paso 3. Examen de diagn√≥stico**

```python
import statsmodels.api as sm
sm.stats.acorr_ljungbox(result.resid, lags=[10], return_df=True)
```

Si el p-valor < 0.05, rechazas $ùêª_0$: hay autocorrelaci√≥n en los residuos ‚Üí el modelo podr√≠a ser mejorado.

Si el p-valor ‚â• 0.05, no hay evidencia de autocorrelaci√≥n ‚Üí los residuos son ruido blanco.

al examinar los errores con la prueba de l jung box el p-valor es mayor que 0.08> 0.05, entonces los errores tienen media igual a cero y varianza constante que significa que no tienen correlaci√≥n entre sus valores en diferentes puntos del tiempo es decir presenta ruido blanco

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np


# Predicciones in-sample (ajustadas)
preds = result.fittedvalues
actual = df["ventas"].dropna()

# Alinear √≠ndices
preds = preds[actual.index]

# M√©tricas
mae = mean_absolute_error(actual, preds)
mse = mean_squared_error(actual, preds)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((actual - preds) / actual)) * 100


# Mostrar resultados
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAPE: {mape:.2f}%")
```

Un MAPE (Mean Absolute Percentage Error) de 8.59% indica que, en promedio, tu modelo de pron√≥stico tiene un error de aproximadamente 8.59% respecto a los valores reales

# **Paso 4. Pron√≥stico**

se hace un pronostico para 20 meses

```python
pred = result.get_forecast(steps=20)
pred_ci = pred.conf_int()

# Graficar
ax = df['ventas'].plot(label='Valor real', figsize=(12, 6))
pred.predicted_mean.plot(ax=ax, label='Pron√≥stico', color='green')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='lightgreen', alpha=0.3)
plt.legend(loc='upper left')
plt.title("Pron√≥stico de las ventas")
plt.show()
```

Entonces, ¬øc√≥mo interpretar los diagn√≥sticos de la trama?

Arriba a la izquierda: los errores residuales parecen fluctuar alrededor de una media de cero y tienen una varianza uniforme.

Arriba a la derecha: la gr√°fica de densidad sugiere una distribuci√≥n normal con media cero.

Abajo a la derecha: el correlograma, tambi√©n conocido como gr√°fico ACF, muestra que los errores residuales no est√°n autocorrelacionados. Cualquier autocorrelaci√≥n implicar√≠a que existe alg√∫n patr√≥n en los errores residuales que no se explican en el modelo. Por lo tanto, deber√° buscar m√°s X (predictores) en el modelo.
